```python
import json
from pymongo import MongoClient
from random import random
import datetime
import random


def getRandDate(interval):
    currentDateTime = datetime.datetime.now()
    randomSeconds = random.randint(0, interval * 24 * 60 * 60)
    randomDateTime = currentDateTime - datetime.timedelta(seconds=randomSeconds)
    return randomDateTime.isoformat()


print(getRandDate(180))
```

```python
from dotenv import dotenv_values

# Load the environment variables from the .env file
env_vars = dotenv_values("../.env")

# Get the values of DB_USER and DB_PASS
user = env_vars.get("DB_USER")
pwd = env_vars.get("DB_PASS")


client = MongoClient('localhost', 27017, username=user, password=pwd)

collection = client['streetvision']['cameras']

data = '/Users/m/Desktop/StreetVisionApp/data_samples/CCTV_and_Traffic_Cameras.geojson'

insert = []
with open(data) as f:
    data = json.load(f)

i =1
for feature in data['features']:
    camItem = {}
    camItem['cam_name'] = 'camera_' + str(i).zfill(4)
    camItem['active_date'] = getRandDate(180)
    camItem['cam_no'] = i
    camItem['is_active'] = random.choice([True, False])
    camItem['provider'] = {'is_residential' : random.choice([True, False]), 'onboard_date': getRandDate(180), 'provider_id':''}
    camItem['location'] = feature['properties']['location']
    camItem['geometry'] = feature['geometry']
    i +=1


    insert.append(camItem)


collection.insert_many(insert)



```
